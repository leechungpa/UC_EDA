```python
import os
import pandas as pd
```

```python
path = "../data/transport/"
file_list = os.listdir(path)

print (file_list)
```

```python
len(file_list)
```

```python
df = pd.DataFrame()
```

```python
for file_name in file_list:
    temp_df = pd.read_excel(path+file_name,sheet_name='Page 6',skiprows=[0,1,5])
    
    trim_temp_df = temp_df.iloc[:,[0,8]].dropna().set_index('Region And State').drop('Subtotal')
    trim_temp_df.columns = [file_name[0:2]+'.'+file_name[2:5]]
    
    df = df.join(trim_temp_df, how='outer')
```

```python
code = pd.read_csv('../data/state_code.csv')
```

```python
final_df = code.set_index('State').join(df).set_index('Code')
```

```python
final_df.to_csv('../data/transport.csv')
```
